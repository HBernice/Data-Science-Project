```python
#Read my dataset into the Jupyter Notebook
import pandas as pd
df=("user_behavior_dataset.csv")
read=pd.read_csv(df)
read
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
read
# Basic scatter plot using Seaborn
plt.figure(figsize=(10,10))
sns.scatterplot(x='Age', y='App Usage Time (min/day)', data=read, hue='Age', size='App Usage Time (min/day)')
plt.title('Scatter Plot of Age vs App Usage Time')
plt.xlabel('Age')
plt.ylabel('App Usage Time (min/day)')
plt.legend(loc='lower right')
plt.show()
```

```python
#INTERPRETATION OF  THE ABOVE SCATTER PLOT
##There is no correlation between the two variables because the points are scattered randomly without any clear pattern.
##The data points are less clustered hence indicating a weak correlation between the two variables.
##No distinct clusters hence indicating no significant patterns in the data.
##There are no points that are far away from the main cluster of data points hence no outliers.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
read
# Set style
sns.set_style("whitegrid")

# pairplot with hue
plt.figure(figsize=(10, 8))
sns.pairplot(read, hue="Operating System", palette="husl")
plt.title('Pairplot with Different Colors based on Operating system')
plt.show()
```

```python
#INTERPRETATION OF THE PAIRPLOT ABOVE
##The diagonal of the pairplot shows a symmetrical bell-shaped curve that indicates the data follows a normal distribution.
##The distributions are either average skewness or skewed to the right hence showing some parts of the data points are concentrated on one side while others are just normally skewed.
##The off diagonal scatter plots show the pairwise relationships between every combination of two numerical variables.
##The clusters that form a straight line indicate positive correlation, meaning they are directly proportional.They also indicate different subsets within the data. The close points also show how strong the correlation relationship is.
##The points that are scattered randomly and do not have any discernible trend indicate that there is likely little correlation between the two variables.
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
read
# Basic histogram using Seaborn
plt.figure(figsize=(8,5))
sns.histplot(read['Data Usage (MB/day)'], bins=25, kde=True, color='orange', edgecolor='black')
plt.title('Histogram of Data Usage Per Day')
plt.xlabel('Data Usage (MB/day)')
plt.ylabel('Frequency')
plt.show()
```

```python
#INTERPRETATION OF THE HISTOGRAM ABOVE
##The distribution is skewed to the right hence showing some data points are concentrated on one side.
##There are no bars that are far from the rest of the distribution hence indicating that there are no outliers.
##The variables takes on a range of 100 to 2500 mb/day.
##Most data points fall in the 100 t0 300 mb/day range. This is indicated by the tall bars.
```

```python
                          #Generating a heatmap

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

read

#Display the first rew rows of the dataset
print("Original Dataframe:")
print(read.head())

#Identify numeric columns
numeric_read=read.select_dtypes(include='number')

#Caalculate the correlation matrix
correlation_matrix=numeric_read.corr()

#Generate the heatmap
plt.figure(figsize=(8,6))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm',
           square=True,linewidths=.5,cbar_kws={"shrink":.8})

#Add Title
plt.title('Correlation Heatmap')

#show the plot
plt.show()
```

```python
#INTERPRETATION OF CORREALATION HEATMAP ABOVE
##Correlation is a statistical measure that describes how two variables are related to each other.
##The heatmap diagonal indicates a perfect positive correlation because of the +1 value. As one variable increases, the other also increases in a perfectly linear way.
## The off diagonal values can also be interpreted. The closer the value is to 1 or -1, the stronger the correlation between the variables indicated.
##The value near 0 means a weak or no linear relationship between the variables.
##The legend shows what each color intensity means in the correlation heatmap. The colors can be used together with the numbers to read out correlation relationships between variables.
## Positive correlations suggest that variables increase or decrease together, while negative correlations indicate that one variable increases as the other decreases.
```
